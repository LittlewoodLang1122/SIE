```python
import numpy as np
from numpy import random as rdm
from matplotlib import pyplot as plt

import utils
import crowd
```

```python
rdm.seed()
c = crowd.crowd(4000, 'con')
c.data
```

```python
c.visualize()
```

```python
x,y = utils.Process(c,rate=0.3, Animator=True)
```

```python
ydata = y[13]
idy_0 = np.argwhere(ydata[2,:] == 0).squeeze()
idy_1 = np.argwhere(ydata[2,:] == 1).squeeze()
idy_2 = np.argwhere(ydata[2,:] == 2).squeeze()
plt.scatter(ydata[0, idy_0], ydata[1, idy_0], c='gold',s=2, label = 'S')
plt.scatter(ydata[0, idy_1], ydata[1, idy_1], c='brown',s=2, label = 'I')
plt.scatter(ydata[0, idy_2], ydata[1, idy_2], c='darkorange',s=2, label = 'R')
plt.legend()
```
